# Modelos lineales y regularizacion

```python
import numpy as np

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
import matplotlib.pyplot as plt

plt.plot(X,y,'b.')
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X] # add x0 = 1 to each instance
w_hat = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
print(w_hat)
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new] # add x0 = 1 to each instance
y_predict = X_new_b.dot(w_hat)
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
print('Coeficientes Sklearn : ',lin_reg.intercept_, lin_reg.coef_)
print('Coeficientes Manual : ',w_hat)
```

# Regularizacion

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * np.sin(X) + 3*np.random.randn(100, 1)
```

```python
import matplotlib.pyplot as plt

plt.plot(X,y,'b.')
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(degree=2,interaction_only=False,include_bias=False)
poly.fit(X)
X_p=poly.transform(X)
```

```python
X_p.shape
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_p, y)
print('Coeficientes Sklearn : ',lin_reg.intercept_, lin_reg.coef_)
```
